### 3. For the pair in Question 2.a plot the scatter plot. Also superimpose a linear regression line on the plot.

image.png

```
# plt.scatter(df['Distance'], df['Time'])
plt.xlabel("Distance")
plt.ylabel("Time")
plt.title("Scatterplot and Linear Regression of Time vs Distance")

# import the Linear Regression class
from sklearn.linear_model import LinearRegression

x = df['Distance']
y = df['Time']

# Choose model parameters

model = LinearRegression(fit_intercept=True)

X = x[:, np.newaxis]

# fit the model to the data
model.fit(X, y)

# Predict label for unknown data
xfit = np.linspace(-1, 81)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)

plt.scatter(x, y)
plt.plot(xfit, yfit, color='r')
plt.show()
```

```
from sklearn.linear_model import Ridge
from yellowbrick.regressor import ResidualsPlot

ridge = Ridge()
visualizer = ResidualsPlot(ridge)
visualizer.fit(X, y)
visualizer.poof()

```
